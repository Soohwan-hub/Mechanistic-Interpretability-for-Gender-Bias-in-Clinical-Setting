## Prepare prompts functions

```python

def prepare_corrupt_prompt(prompt:str ) -> Dict[str, Any]:
  

    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": prompt}
    ]

    corrupted_prompt = llm.tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
    )
    
  

    corrupted_tokens = llm.tokenizer(
        corrupted_prompt,
        return_tensors="pt",
    )["input_ids"][0]

    return {
        "corrupted_tokens": corrupted_tokens,
        "corrupted_prompt": corrupted_prompt,
    }

def prepare_clean_prompt(target_gender: str) -> Dict[str, Any]:
    """Prepare the clean prompt and locate the gender token position."""

    message = {
        "role": "user",
        "content": f"The patient is {target_gender}",
    }

    clean_prompt = llm.tokenizer.apply_chat_template(
        [message],
        tokenize=False,
        add_generation_prompt=True,
    )

    # `pt` returns PyTorch tensors.
    # `input_ids` is a sequence of integer token IDs of shape (batch, seq_len).
    # [0] selects the first batch -> (seq_len,).
    clean_tokens = llm.tokenizer(clean_prompt, return_tensors="pt")["input_ids"][0]

    print("\nClean tokens:    ", [llm.tokenizer.decode(t) for t in clean_tokens])

    # Get the token ids of the target gender only.
    target_token_ids = llm.tokenizer(
        " " + target_gender,
        return_tensors="pt",
        add_special_tokens=False,
    )["input_ids"][0]

    # The last token of `target_gender` is the gender token.
    # We get the position where the clean tokens match the last target_gender token.
    # [0] gets the first occurrence of match, [0] unwraps the single-element tensor
    # to get the actual integer value; [0][0] leaves the tensor 0-dim. Therefore,
    # `.tolist()` returns a scalar, not a list. `patch_token_from` should be a
    # scalar value to use it for indexing later.
    patch_token_from = torch.argwhere(clean_tokens == target_token_ids[-1])[0][0].tolist()

    return {
        "clean_prompt": clean_prompt,
        "clean_tokens": clean_tokens,
        "patch_token_from": patch_token_from,
    }

```

## Clean activations

```python
def resolve(proxy):
    """
    Safely unwrap an nnsight saved proxy.
    In some nnsight versions the saved tensor is at proxy.value;
    in others the proxy itself becomes the tensor after the context exits.
    """
    if hasattr(proxy, 'value'):
        return proxy.value
    return proxy
# ── Pre-compute clean prompt & activations ONCE (shared across all 45 runs) ──
clean_prompt_output = prepare_clean_prompt(target_gender)

clean_prompt        = clean_prompt_output["clean_prompt"]
clean_tokens        = clean_prompt_output["clean_tokens"]
patch_token_from    = clean_prompt_output["patch_token_from"]

_saved_clean: Dict[int, Any] = {}
with torch.no_grad():
    with llm.generate(max_new_tokens=1) as tracer:
        with tracer.invoke(clean_prompt):
            for layer_idx in range(num_layers_model):
                act = llm.model.layers[layer_idx].output[0]
                _saved_clean[layer_idx] = act[patch_token_from, :].save()

clean_activations_cache: Dict[int, torch.Tensor] = {
    i: resolve(proxy).detach().clone() for i, proxy in _saved_clean.items()
}

print(f"Cached {len(clean_activations_cache)} layers")
print(f"Vector shape per layer : {clean_activations_cache[0].shape}")
```

```python
# ── PATCHING FUNCTION (receives already-filled prompt) ───────────────────────

def _patch_one_run(
    prompt: str,
    run_save_dir: str,
    num_layers_model :int ,
num_layers_batch  : int = 3
) -> tuple[np.ndarray, list[str], dict]:
    """
    Receives a fully-filled prompt (BHC and CONDITION_NAME already substituted).
    Runs the full residual-stream patching loop and returns:
      - rewrite_matrix : (num_layers, num_tokens) np.ndarray
      - token_labels   : list[str]
      - layer_hidden_states : dict[int, np.ndarray]  shape (num_tokens, d_model) per layer
    Results are cached layer-by-layer to disk for crash recovery.
    """
    os.makedirs(run_save_dir, exist_ok=True)

    # Build corrupted prompt from the already-filled clean prompt
    corrupt_out       = prepare_corrupt_prompt(prompt)
    corrupted_prompt  = corrupt_out["corrupted_prompt"]
    corrupted_tokens  = corrupt_out["corrupted_tokens"]
    num_tokens        = len(corrupted_tokens)

    diff   = len(clean_tokens) - num_tokens
    offset = diff if diff > 0 else 0

    # Corrupted baseline (no patching)
    with torch.no_grad():
        with llm.generate(max_new_tokens=3) as tracer:
            with tracer.invoke(corrupted_prompt):
                corrupted_logits = llm.lm_head.next().next().output
                _target_prob = softmax(corrupted_logits[0, 0, :])[target_gender_token].save()
    corrupted_prob_value_target = resolve(_target_prob).cpu().float().item()

    # Patching loop
    all_layer_scores: list[list[float]] = []
    layer_hidden_states: dict[int, np.ndarray] = {}   #  hidden states per layer

    for layer_idx in range(0, num_layers_model, num_layers_batch):
        end        = min(layer_idx + num_layers_batch, num_layers_model)
        batch_path = os.path.join(run_save_dir, f"layers_{layer_idx}_{end-1}.pkl")
    
        if os.path.exists(batch_path):       # crash recovery
            with open(batch_path, "rb") as f:
                batch = pickle.load(f)
            all_layer_scores.extend(batch["rewrite_scores"])
            layer_hidden_states.update(batch.get("hidden_states", {}))
            continue

        batch_scores: list[list[float]] = []
        batch_hidden: dict[int, np.ndarray] = {}   

        for l in range(layer_idx, end):
            print(f"Layer{layer_idx}")
            layer_scores: list[float] = []

            #  capture hidden state at layer l on the corrupted prompt (no patching)
            
            torch.cuda.empty_cache()
            with torch.no_grad():
                with llm.generate(max_new_tokens=3) as tracer:
                    with tracer.invoke(corrupted_prompt):
                        hs = llm.model.layers[l].next().next().output[0].save()   # (seq_len, d_model)
            batch_hidden[l] = resolve(hs).cpu().float().numpy()

            for token_idx in range(num_tokens):
                torch.cuda.empty_cache()
                with torch.no_grad():
                    with llm.generate(max_new_tokens=3) as tracer:
                        with tracer.invoke(corrupted_prompt):
                            act = llm.model.layers[l].output[0]
                            act[token_idx + offset, :] = clean_activations_cache[l]
                            llm.model.layers[l].output[0] = act
                            patched_logits = llm.lm_head.next().next().output
                            patched_prob   = softmax(
                                patched_logits[0, 0, :]
                            )[target_gender_token].save()
                p = resolve(patched_prob).cpu().float().item()
                rewrite_score = (p - corrupted_prob_value_target) / (
                    1 - corrupted_prob_value_target 
                )
                layer_scores.append(rewrite_score)
            batch_scores.append(layer_scores)

        with open(batch_path, "wb") as f:
            pickle.dump({
                "rewrite_scores": batch_scores,
                "hidden_states":  batch_hidden,   
                "start": layer_idx,
                "end":   end,
            }, f)
        all_layer_scores.extend(batch_scores)
        layer_hidden_states.update(batch_hidden)

    token_labels = [llm.tokenizer.decode([t]) for t in corrupted_tokens]
    return (
        np.array(all_layer_scores),   # (num_layers, num_tokens)
        token_labels,                  # list[str]
        layer_hidden_states,           # dict[int -> (seq_len, d_model)]
    )


# ── MAIN LOOP ────────────────────────────────────────────────────────────────

prompt_types = {
    #"VIGNETTE_PROMPT_A": VIGNETTE_PROMPT_A,
    #"VIGNETTE_PROMPT_B": VIGNETTE_PROMPT_B,
    "VIGNETTE_PROMPT_C": VIGNETTE_PROMPT_C,
}
prompt_vars = [f'var{i}' for i in range(1, 6)]   # 5 prompt variations

def run(condition_name, condition_cases):
    for case in condition_cases:
        for prompt_name, prompt_t in prompt_types.items():
            for var in prompt_vars:
                run_save_dir = os.path.join(BASE_SAVE_DIR, prompt_name, var)
                os.makedirs(run_save_dir, exist_ok=True)
                run_path = os.path.join(run_save_dir, "result.pkl")

                if os.path.exists(run_path):
                    print(f"Skipping {prompt_name}/{var} — already exists")
                    continue

                print(f"Running {prompt_name}/{var} ...")

                # Fill placeholders HERE in the main loop, pass filled prompt down
                filled_prompt = (
                    prompt_t[var]
                    .replace("[BHC]", case)
                    .replace("[CONDITION_NAME]", condition_name)
                )

                rewrite_matrix, token_labels, layer_hidden_states = _patch_one_run(
                    prompt=filled_prompt,
                    run_save_dir=run_save_dir,
                    num_layers_model  = len(llm.model.layers),
                    num_layers_batch  = 3
                )

                with open(run_path, "wb") as f:
                    pickle.dump({
                        "rewrite_matrix":     rewrite_matrix,      # (num_layers, num_tokens)
                        "token_labels":       token_labels,         # list[str]
                        "layer_hidden_states": layer_hidden_states, # dict[int -> np.ndarray]
                        "prompt_name":        prompt_name,
                        "var":                var,
                    }, f)
                print(f"  ✓ Saved {prompt_name}/{var}")
```

```python
# ── HIDDEN STATE CAPTURE AT GENERATION STEP 3 (Female/Male step) ─────────────

def capture_hidden_states_step3(
    prompt: str,
    num_layers_model: int,
) -> dict[int, np.ndarray]:
    """
    Captures residual stream hidden states at generation step 3 (Female/Male)
    for all layers, on the corrupted prompt without any patching.
    Returns dict[layer_idx -> np.ndarray of shape (seq_len, d_model)]
    """
    corrupt_out      = prepare_corrupt_prompt(prompt)
    corrupted_prompt = corrupt_out["corrupted_prompt"]

    hidden_states = {}

    for l in range(num_layers_model):
        torch.cuda.empty_cache()
        with torch.no_grad():
            with llm.generate(max_new_tokens=3) as tracer:
                with tracer.invoke(corrupted_prompt):
                    hs = llm.model.layers[l].next().next().output[0].save()
        hidden_states[l] = resolve(hs).cpu().float().numpy()
        print(f"  Layer {l:>2} captured — shape: {hidden_states[l].shape}")

    return hidden_states


# ── TEST: run on one example ──────────────────────────────────────────────────
test_prompt = (
    VIGNETTE_PROMPT_A["var1"]
    .replace("[BHC]", condition_cases[0])       # first case
    .replace("[CONDITION_NAME]", condition_name)
)

print("Capturing hidden states at generation step 3 (Female/Male)...")
hs_step3 = capture_hidden_states_step3(
    prompt           = test_prompt,
    num_layers_model = len(llm.model.layers),
)
print(f"\nDone. Captured {len(hs_step3)} layers.")
print(f"Shape at layer 18: {hs_step3[18].shape}")
```

```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt

# ── CONFIG ────────────────────────────────────────────────────────────────────
BASE_DIR    = "/home/ubuntu"
PROMPT_DIRS = {"PROMPT_A": "PROMPTA", "PROMPT_B": "PROMPTB", "PROMPT_C": "PROMPTC"}
VAR_NUMS    = range(1, 6)
N_LAYERS    = 28

COLORS      = {"PROMPT_A": "#2196F3", "PROMPT_B": "#FF5722", "PROMPT_C": "#4CAF50"}

# ═════════════════════════════════════════════════════════════════════════════
# 1.  LOAD
# ═════════════════════════════════════════════════════════════════════════════
def load_results(base_dir, prompt_dirs, var_nums):
    """
    Returns
    -------
    data : dict  prompt_name ->
                   { "matrices":      list of (28, n_tokens) arrays,
                     "token_labels":  list of label-lists (one per variation) }
    """
    data = {}
    for prompt_name, prompt_dir in prompt_dirs.items():
        matrices, token_labels = [], []
        for var_num in var_nums:
            path = os.path.join(base_dir, prompt_dir, f"var{var_num}", "result.pkl")
            if not os.path.exists(path):
                print(f"  SKIP (not found): {path}")
                continue
            try:
                with open(path, "rb") as f:
                    r = pickle.load(f)
            except Exception as e:
                print(f"  SKIP (load error) {path}: {e}")
                continue

            matrix = np.array(r["rewrite_matrix"])   # (28, n_tokens)
            matrices.append(matrix)
            token_labels.append(r["token_labels"])
            print(f"  Loaded {prompt_name}/VAR{var_num}  shape={matrix.shape}")

        if matrices:
            data[prompt_name] = {"matrices": matrices, "token_labels": token_labels}
            print(f"  → {prompt_name}: {len(matrices)} variations loaded\n")
        else:
            print(f"  WARNING: no data for {prompt_name}\n")
    return data


# ═════════════════════════════════════════════════════════════════════════════
# 2.  AGGREGATE
# ═════════════════════════════════════════════════════════════════════════════
def aggregate(data):
    """
    Returns
    -------
    layer_means  : prompt_name -> (28,)   mean rewrite score per layer
    layer_stds   : prompt_name -> (28,)   std across variations
    token_means  : prompt_name -> (min_len,)  mean rewrite score per token position
    token_labels : prompt_name -> list[str],  len = min_len
    
    Aggregation rationale
    ─────────────────────
    Layer axis  : matrix.mean(axis=1) collapses the token dimension → (28,).
                  Comparing this scalar across variations is valid even when
                  variations differ in token count, because we are averaging
                  *within* each run before comparing *across* runs.

    Token axis  : matrix.mean(axis=0) collapses the layer dimension → (n_tokens,).
                  Variations differ in total token count (condition names vary
                  in length), but the suffix ending at "Gender:" is stable.
                  We therefore right-align — keeping the last min_len tokens —
                  so that position 0 of every truncated vector corresponds to
                  the same suffix-relative token position.
    """
    layer_means, layer_stds = {}, {}
    token_means, token_labels_out = {}, {}

    for prompt_name, d in data.items():
        matrices     = d["matrices"]
        labels_list  = d["token_labels"]

        # ── per-layer aggregation ─────────────────────────────────────────────
        per_layer_vectors = [m.mean(axis=1) for m in matrices]   # each (28,)
        stacked = np.stack(per_layer_vectors)                      # (n_vars, 28)
        layer_means[prompt_name] = stacked.mean(axis=0)
        layer_stds[prompt_name]  = stacked.std(axis=0)

        # ── per-token aggregation (right-aligned) ─────────────────────────────
        per_token_vectors = [m.mean(axis=0) for m in matrices]    # each (n_tokens,)
        min_len = min(len(v) for v in per_token_vectors)

        # right-align: take the last min_len positions so that the
        # "Gender:" token (always last) is always at index min_len-1
        truncated = [v[-min_len:] for v in per_token_vectors]
        token_means[prompt_name] = np.stack(truncated).mean(axis=0)  # (min_len,)

        # use label list from the shortest variation (it set the floor)
        shortest_idx = int(np.argmin([len(v) for v in per_token_vectors]))
        token_labels_out[prompt_name] = labels_list[shortest_idx][-min_len:]

        print(f"  {prompt_name}: layer shape={layer_means[prompt_name].shape}, "
              f"token shape={token_means[prompt_name].shape}, "
              f"last label={token_labels_out[prompt_name][-1]}")

    return layer_means, layer_stds, token_means, token_labels_out


# ═════════════════════════════════════════════════════════════════════════════
# 3.  PLOTS
# ═════════════════════════════════════════════════════════════════════════════

def plot_layers(layer_means, layer_stds, save_path):
    """
    Plot 1 — Per-layer mean rewrite score (±1 std band across variations).
    X axis : layer index 0-27
    Y axis : mean rewrite score averaged across all token positions and variations
    Band   : ±1 std across variations (captures prompt-variation uncertainty)
    Peak markers show which layer carries the most signal per prompt type.
    """
    layers = np.arange(N_LAYERS)
    fig, ax = plt.subplots(figsize=(12, 5))

    for prompt_name, mean in layer_means.items():
        std   = layer_stds[prompt_name]
        color = COLORS.get(prompt_name, "grey")
        ax.plot(layers, mean, label=prompt_name, color=color, linewidth=2.0)
        ax.fill_between(layers, mean - std, mean + std, color=color, alpha=0.15)

        peak_layer = int(np.argmax(mean))
        ax.axvline(peak_layer, color=color, linewidth=0.8, linestyle="--", alpha=0.5)
        ax.annotate(f"L{peak_layer}",
                    xy=(peak_layer, mean[peak_layer]),
                    xytext=(4, 4), textcoords="offset points",
                    fontsize=8, color=color)

    ax.axhline(0, color="black", linewidth=0.6, linestyle="--", alpha=0.4)
    ax.set_xlabel("Layer", fontsize=11)
    ax.set_ylabel("Mean rewrite score\n(avg across tokens & variations)", fontsize=11)
    ax.set_title("Plot 1 — Per-layer mean rewrite score\n"
                 "shaded band = ±1 std across prompt variations", fontsize=12)
    ax.set_xticks(layers)
    ax.legend(fontsize=10)
    ax.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    print(f"Saved: {save_path}")


def plot_tokens(token_means, token_labels, save_path):
    """
    Plot 2 — Per-token mean rewrite score (diagnostic: boilerplate check).
    X axis : token position in the right-aligned suffix (position -1 = Gender:)
    Y axis : mean rewrite score averaged across all layers and variations
    Dotted vertical line marks the Gender: token for each prompt.

    If boilerplate tokens (system prompt, filler words) show high scores,
    that would suggest the patching signal is diffuse and not specific to
    the clinical condition / demographic token — a red flag for interpretation.
    The expected result is that signal concentrates near the end of the prompt.
    """
    fig, ax = plt.subplots(figsize=(24, 5))

    for prompt_name, per_token_mean in token_means.items():
        labels = token_labels[prompt_name]
        x      = np.arange(len(per_token_mean))
        color  = COLORS.get(prompt_name, "grey")

        ax.plot(x, per_token_mean, label=prompt_name,
                color=color, linewidth=1.4, alpha=0.85)

        # "Gender" may appear as "Gender_47" etc. — match by substring
        gender_positions = [i for i, t in enumerate(labels) if "Gender" in t]
        for gp in gender_positions:
            ax.axvline(gp, color=color, linewidth=0.8, linestyle=":", alpha=0.6)

    # x-axis ticks from the first prompt's labels
    ref_labels = next(iter(token_labels.values()))
    n_ref      = len(ref_labels)
    tick_step  = max(1, n_ref // 40)
    ax.set_xticks(np.arange(n_ref)[::tick_step])
    ax.set_xticklabels(ref_labels[::tick_step], rotation=90, fontsize=7)

    ax.axhline(0, color="black", linewidth=0.6, linestyle="--", alpha=0.4)
    ax.set_xlabel("Token position (right-aligned suffix)", fontsize=11)
    ax.set_ylabel("Mean rewrite score\n(avg across 28 layers & variations)", fontsize=11)
    ax.set_title("Plot 2 — Per-token mean rewrite score\n"
                 "dotted line = Gender: token | checks if boilerplate drives signal",
                 fontsize=12)
    ax.legend(fontsize=10)
    ax.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.show()
    print(f"Saved: {save_path}")


def plot_top_tokens(token_means, token_labels, save_path, top_n=20):
    """
    Plot 3 — Top-N tokens by mean rewrite score (one bar chart per prompt).
    Each subplot ranks token positions by their mean rewrite score (descending).
    This surfaces which specific tokens — condition name, demographic marker,
    punctuation — are most causally implicated in the model's gender prediction,
    independent of their position in the sequence.
    """
    n_prompts = len(token_means)
    fig, axes = plt.subplots(1, n_prompts, figsize=(8 * n_prompts, 6), sharey=False)
    if n_prompts == 1:
        axes = [axes]

    for ax, (prompt_name, per_token_mean) in zip(axes, token_means.items()):
        labels = token_labels[prompt_name]
        color  = COLORS.get(prompt_name, "grey")

        top_idx    = np.argsort(per_token_mean)[::-1][:top_n]
        top_scores = per_token_mean[top_idx]
        top_labels = [labels[i] for i in top_idx]

        ax.barh(range(top_n), top_scores[::-1], color=color, alpha=0.8)
        ax.set_yticks(range(top_n))
        ax.set_yticklabels(top_labels[::-1], fontsize=9)
        ax.axvline(0, color="black", linewidth=0.6, linestyle="--", alpha=0.4)
        ax.set_xlabel("Mean rewrite score", fontsize=10)
        ax.set_title(f"{prompt_name}\nTop {top_n} tokens by rewrite score", fontsize=11)
        ax.grid(axis="x", alpha=0.3)

    plt.suptitle("Plot 3 — Top tokens by mean rewrite score\n"
                 "(avg across 28 layers & variations)", fontsize=12, y=1.02)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150, bbox_inches="tight")
    plt.show()
    print(f"Saved: {save_path}")


# ═════════════════════════════════════════════════════════════════════════════
# MAIN
# ═════════════════════════════════════════════════════════════════════════════
data = load_results(BASE_DIR, PROMPT_DIRS, VAR_NUMS)

layer_means, layer_stds, token_means, token_labels = aggregate(data)

plot_layers(
    layer_means, layer_stds,
    save_path=os.path.join(BASE_DIR, "plot1_per_layer.png")
)

plot_tokens(
    token_means, token_labels,
    save_path=os.path.join(BASE_DIR, "plot2_per_token.png")
)

plot_top_tokens(
    token_means, token_labels,
    save_path=os.path.join(BASE_DIR, "plot3_top_tokens.png"),
    top_n=20
)
```

```python
import pickle
import numpy as np
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import os

# ── CONFIG ────────────────────────────────────────────────────────────────────
BASE_DIR     = "/home/ubuntu"
PROMPT_DIRS  = {"PROMPT_C": "PROMPTC", "PROMPT_B": "PROMPTB"}
VAR_NUMS     = [1, 2, 3]
TARGET_LAYER = 27
TOP_K        = 20
# positions to inspect: we'll auto-detect + let you add specific indices
OUTPUT_PATH  = os.path.join(BASE_DIR, "logit_lens_topk_layer18.txt")

# ── LOAD STATIC WEIGHTS ───────────────────────────────────────────────────────
from transformers import AutoModelForCausalLM

raw_model   = AutoModelForCausalLM.from_pretrained(
    "Qwen/Qwen2.5-7B-Instruct",
    torch_dtype=torch.float32,
    device_map="cpu",
    low_cpu_mem_usage=True,
)
W_U         = raw_model.lm_head.weight.detach().float().cpu()    # (vocab_size, 3584)
norm_weight = raw_model.model.norm.weight.detach().float().cpu() # (3584,)
norm_eps    = raw_model.model.norm.variance_epsilon
tokenizer   = raw_model.config  # we'll use llm.tokenizer instead — see below

del raw_model
torch.cuda.empty_cache()
print("Weights extracted.")

# ── RMS NORM ──────────────────────────────────────────────────────────────────
def rms_norm(x, weight, eps):
    variance = x.pow(2).mean(-1, keepdim=True)
    x = x * torch.rsqrt(variance + eps)
    return weight * x

# ── TOP-K LOGIT LENS AT ONE POSITION ─────────────────────────────────────────
def top_k_at_position(h_layer, token_pos, k=TOP_K):
    """
    h_layer : np.ndarray (n_tokens, 3584) — hidden states at target layer
    token_pos: int — which token position to inspect
    returns  : list of (token_str, probability) tuples
    """
    h      = torch.tensor(h_layer[token_pos], dtype=torch.float32)  # (3584,)
    h_norm = rms_norm(h, norm_weight, norm_eps)                      # (3584,)
    logits = W_U @ h_norm                                            # (vocab_size,)
    probs  = F.softmax(logits, dim=0)                                # (vocab_size,)

    top_probs, top_ids = torch.topk(probs, k)
    decoded = [llm.tokenizer.decode([tid.item()]) for tid in top_ids]
    return list(zip(decoded, top_probs.tolist()))

# ── FIND INTERESTING POSITIONS ────────────────────────────────────────────────
def get_positions_of_interest(token_labels):
    """
    Auto-detect: Gender: token, last token, and condition-related tokens.
    Returns dict of {label: index}
    """
    positions = {}
    for i, t in enumerate(token_labels):
        if t == "Gender":
            positions["Gender:"] = i
        if t in ("Depression"):
            positions[f"condition({t})"] = i
    # always include last token
    positions["last_token"] = len(token_labels) - 1
    return positions

# ── MAIN LOOP ─────────────────────────────────────────────────────────────────
report_lines = []

for prompt_name, prompt_dir in PROMPT_DIRS.items():
    report_lines.append(f"\n{'='*70}")
    report_lines.append(f"  {prompt_name} — Logit Lens at Layer {TARGET_LAYER}")
    report_lines.append(f"{'='*70}")

    for var_num in VAR_NUMS:
        path = os.path.join(BASE_DIR, prompt_dir, f"var{var_num}", "result.pkl")
        with open(path, "rb") as f:
            r = pickle.load(f)

        h_layer      = r["layer_hidden_states"][TARGET_LAYER]  # (n_tokens, 3584)
        token_labels = r["token_labels"]
        positions    = get_positions_of_interest(token_labels)

        report_lines.append(f"\n  --- var{var_num} ---")
        report_lines.append(f"  Token labels (last 10): {token_labels[-10:]}\n")

        for pos_name, pos_idx in positions.items():
            top_k = top_k_at_position(h_layer, pos_idx)

            report_lines.append(f"  Position {pos_idx:>4} [{pos_name:>20}]  "
                                 f"token='{token_labels[pos_idx]}'")
            for rank, (tok, prob) in enumerate(top_k):
                report_lines.append(f"      {rank+1:>2}.  {repr(tok):>20}  {prob:.4f}")
            report_lines.append("")

# ── SAVE & PRINT ──────────────────────────────────────────────────────────────
report = "\n".join(report_lines)
print(report)

with open(OUTPUT_PATH, "w") as f:
    f.write(report)
print(f"\nSaved to: {OUTPUT_PATH}")
```
